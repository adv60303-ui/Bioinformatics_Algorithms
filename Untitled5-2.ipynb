```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.spatial import distance

%matplotlib inline
```

```python
img = plt.imread("flower.png")[:, :, :3]
H, W, C = img.shape
X = img.reshape(-1, C).astype(np.float64)

print(img.shape, X.shape)
```

```python
class KMeans:
    def __init__(self, n_clusters, max_iter, seed=0, tol=1e-3):
        self.n_clusters = int(n_clusters)
        self.max_iter = int(max_iter)
        self.seed = int(seed)
        self.tol = float(tol)
        self.centroids = None
        self.losses = []

    def distances(self, X):
        if self.centroids is None:
            raise ValueError("centroids are not initialized")
        return distance.cdist(X, self.centroids, metric="euclidean")

    def loss(self, dist):
        min_d = np.min(dist, axis=1)
        return float(np.sum(min_d ** 2))

    def initialize(self, X):
        n = X.shape[0]
        if self.n_clusters > n:
            raise ValueError("n_clusters must be <= number of samples")
        idx = np.random.choice(n, size=self.n_clusters, replace=False)
        self.centroids = X[idx].copy()

    def fit_predict(self, X):
        np.random.seed(self.seed)
        self.initialize(X)
        self.losses = []

        for _ in range(self.max_iter):
            dist = self.distances(X)
            labels = np.argmin(dist, axis=1)
            self.losses.append(self.loss(dist))

            old = self.centroids.copy()
            new = old.copy()

            for k in range(self.n_clusters):
                mask = labels == k
                if not np.any(mask):
                    new[k] = X[np.random.randint(0, X.shape[0])]
                else:
                    new[k] = X[mask].mean(axis=0)

            self.centroids = new
            if np.allclose(old, new, atol=self.tol):
                break

        return np.argmin(self.distances(X), axis=1)
```

```python
def compress_image(img, k, max_iter=100, seed=0, tol=1e-3):
    H, W, C = img.shape
    X = img.reshape(-1, C).astype(np.float64)

    km = KMeans(n_clusters=k, max_iter=max_iter, seed=seed, tol=tol)
    labels = km.fit_predict(X)
    centroids = km.centroids

    Xc = centroids[labels]
    img_c = Xc.reshape(H, W, C)
    if np.issubdtype(img.dtype, np.integer):
        img_c = np.clip(np.rint(img_c), 0, 255).astype(img.dtype)
    else:
        img_c = np.clip(img_c, 0.0, 1.0).astype(img.dtype)

    return img_c, km


Ks = [2, 4, 8, 16]
fig, axes = plt.subplots(1, len(Ks) + 1, figsize=(4 * (len(Ks) + 1), 4))
axes[0].imshow(img)
axes[0].set_title("original")
axes[0].axis("off")

for i, k in enumerate(Ks, start=1):
    img_c, km = compress_image(img, k=k, max_iter=50, seed=0)
    axes[i].imshow(img_c)
    axes[i].set_title(f"k={k}")
    axes[i].axis("off")

plt.show()

plt.figure(figsize=(5, 3))
plt.plot(km.losses)
plt.xlabel("iteration")
plt.ylabel("loss")
plt.title(f"KMeans loss (k={Ks[-1]})")
plt.grid(True)
plt.show()
```

```python
def distance_matrix(X):
    X = np.asarray(X, dtype=np.float64)
    dif = X[:, None, :] - X[None, :, :]
    return np.linalg.norm(dif, axis=2)


def linkage(D):
    D = np.asarray(D, dtype=np.float64)
    n = D.shape[0]


    clusters = {i: [i] for i in range(n)}
    active = list(range(n))
    Z = []
    next_id = n

    for _ in range(n - 1):
        best = None
        best_d = np.inf

        for ai in range(len(active)):
            for aj in range(ai + 1, len(active)):
                c1 = active[ai]
                c2 = active[aj]
                pts1 = clusters[c1]
                pts2 = clusters[c2]
                d = D[np.ix_(pts1, pts2)].mean()
                if d < best_d:
                    best_d = d
                    best = (c1, c2)

        c1, c2 = best
        new_pts = clusters[c1] + clusters[c2]
        Z.append([float(c1), float(c2), float(best_d), float(len(new_pts))])

        clusters[next_id] = new_pts
        active = [c for c in active if c not in (c1, c2)]
        active.append(next_id)
        next_id += 1

    return np.asarray(Z, dtype=float)
```

```python
D = np.array([[0.  , 0.74, 0.85, 0.54, 0.83, 0.92, 0.89],
              [0.74, 0.  , 1.59, 1.35, 1.2 , 1.48, 1.55],
              [0.85, 1.59, 0.  , 0.63, 1.13, 0.69, 0.73],
              [0.54, 1.35, 0.63, 0.  , 0.66, 0.43, 0.88],
              [0.83, 1.2 , 1.13, 0.66, 0.  , 0.72, 0.55],
              [0.92, 1.48, 0.69, 0.43, 0.72, 0.  , 0.8 ],
              [0.89, 1.55, 0.73, 0.88, 0.55, 0.8 , 0.  ]])

Z = linkage(D)
print(Z)
```

```python
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram

rng = np.random.default_rng(0)
digits = datasets.load_digits().images

digits = rng.permutation(digits, axis=0)[:20]

XX = digits.reshape((digits.shape[0], -1))

D_digits = distance_matrix(XX)
Z_digits = linkage(D_digits)

plt.figure(figsize=(10, 4))
dendrogram(Z_digits)
plt.title("Hierarchical clustering (average linkage) — digits")
plt.xlabel("sample")
plt.ylabel("distance")
plt.show()
```
